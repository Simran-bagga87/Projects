## Linear regression model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X = df.drop(['Monthly_Revenue'],axis =1)
y = df['Monthly_Revenue']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
predicted_value = lm.predict(X_test)
```

```python
sns.scatterplot(x=predicted_value,y=y_test)

```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("THE ROOT MEAN SQUARED ERROR OF LINEAR REGRESSION MODEL IS: ",np.sqrt(mean_squared_error(predicted_value,y_test)).round(2))
```

```python

pd.DataFrame(data=lm.coef_,index=df.columns[0:7],columns=['COEFFICIENT'])
```

##### This shows that increase in the Marketing Spend by 1 unit and other features fixed lead to 4.94 increase in monthy revenue which is maximum out of other features


## Random forest Regressor (RFR)

```python
from sklearn.ensemble import RandomForestRegressor
rfm = RandomForestRegressor(n_estimators=1)

#elbow method
errorrate = []
for i in range(1,600,50):
    rfm = RandomForestRegressor(n_estimators=i)
    rfm.fit(X_train,y_train)
    random_f_pred = rfm.predict(X_test)
    errorrate.append(mean_absolute_error(random_f_pred,y_test))

```

```python

sns.lineplot(y=errorrate,x=range(1,600,50),marker='o',markerfacecolor='red')
plt.title("Elbow method predicting best number of trees for optimal solution")
plt.ylabel('Error rate')
```

```python
rfm = RandomForestRegressor(n_estimators=350)
rfm.fit(X_train,y_train)
random_f_pred = rfm.predict(X_test)
```

```python
sns.scatterplot(x=random_f_pred,y=y_test)
plt.title("Random Forest Predicted value vs actual value")
plt.xlabel('Predicted Value')
plt.ylabel('Actual value')
```

```python
print("THE ROOT MEAN SQUARED ERROR OF RANDOM FOREST IS: ",np.sqrt(mean_squared_error(random_f_pred,y_test)).round(2))
```

## XG BOOST MODEL


```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],     #number of trees
    'learning_rate': [0.01, 0.1, 0.2],    #increment in weight to reduce previous error
    'max_depth': [3, 5, 7],               #the depth of tree for complex computation(lead to overfit)
    'subsample': [0.8, 0.9, 1.0]          #fraction of sample tree increase lead to overfit 
}
xgbmodel =xgb.XGBRegressor(objective='reg:squarederror')
gs = GridSearchCV(xgbmodel,param_grid,verbose=3)
```

```python
gs.fit(X_train,y_train)
```

```python
bestmodel = gs.best_estimator_
```

```python
bestmodel.fit(X_train,y_train)
```

```python
XGB_pred = bestmodel.predict(X_test)
```

```python
sns.scatterplot(x=XGB_pred,y=y_test)
plt.title("XGBoost Model Predicted value vs actual value")
plt.xlabel('Predicted Value')
plt.ylabel('Actual value')
```

```python
print("THE ROOT MEAN SQUARED ERROR OF XGBOOST MODEL IS: ",np.sqrt(mean_squared_error(XGB_pred,y_test)).round(2))
```
